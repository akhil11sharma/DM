# Importing Library

```python
import pandas as pd
import numpy as np
```

# Reading File

```python
data=pd.read_excel("/content/Data.xlsx")
```

```python
x=data['Sales of toys']
y=data['Sales of General store']
```

```python
mean_x=x.mean()
mean_y=y.mean()
```

```python
median_x=x.median()
median_y=y.median()
```

```python
mode_x=x.mode().iloc[0]
mode_y=y.mode().iloc[0]
```

# using scipy
